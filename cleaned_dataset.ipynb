```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("merged_hospital_data.csv")
```

```python
data["Transfer_Date"]
```

```python
data["Insurance Provider"] = data["Insurance Provider"].fillna(data["Insurance Provider"].mode()[0])
```

```python
data["Readmission"] = data["Readmission"].fillna(data["Readmission"].mode()[0])
```

```python
data["Dept_Bed_Capacity_Derived"] = data["Dept_Bed_Capacity_Derived"].fillna(data["Dept_Bed_Capacity_Derived"].mean())
```

```python
data["Beds_Occupied_Count"] = data["Beds_Occupied_Count"].fillna(data["Beds_Occupied_Count"].mode()[0])
```

```python
data["Equipment ID"] = data["Equipment ID"].fillna(data["Equipment ID"].mode()[0])
```

```python
data["Equipment Number"] = data["Equipment Number"].fillna(data["Equipment Number"].mode()[0])
```

```python
data["Equipment Status"] = data["Equipment Status"].fillna(data["Equipment Status"].mode()[0])
```

```python
data["Equipment_Total_Inventory"] = data["Equipment_Total_Inventory"].fillna(data["Equipment_Total_Inventory"].mean())
```

```python
data["Equipment_Usage_Duration_Hours"] = data["Equipment_Usage_Duration_Hours"].fillna(data["Equipment_Usage_Duration_Hours"].mean())
```

```python
data["Transferred"] = data["Transferred"].fillna(data["Transferred"].mode()[0])
```

```python
data["Transfer_From_Department"] = data["Transfer_From_Department"].fillna(data["Transfer_From_Department"].mode()[0])
```

```python
data["Transfer_To_Department"] = data["Transfer_To_Department"].fillna(data["Transfer_To_Department"].mode()[0])
```

```python
data["Number_of_Transfers"] = data["Number_of_Transfers"].fillna(data["Number_of_Transfers"].mean())
```

```python
data["Dept_ICU_Bed_Capacity_Derived"] = data["Dept_ICU_Bed_Capacity_Derived"].fillna(data["Dept_ICU_Bed_Capacity_Derived"].mean())
```

```python
data["Dept_Staff_Capacity_Derived"] = data["Dept_Staff_Capacity_Derived"].fillna(data["Dept_Staff_Capacity_Derived"].mean())
```

```python
data["Admissions_Rate_%_Derived"] = data["Admissions_Rate_%_Derived"].fillna(data["Admissions_Rate_%_Derived"].mean())
```

```python
data["Staff_Utilization_%_Derived"] = data["Staff_Utilization_%_Derived"].fillna(data["Staff_Utilization_%_Derived"].mean())
```

```python
data["Bed_Occupancy_Rate_%"] = data["Bed_Occupancy_Rate_%"].fillna(data["Bed_Occupancy_Rate_%"].mean())
```

```python
data["Bed_Occupancy_Rate_Calc"] = data["Bed_Occupancy_Rate_Calc"].fillna(data["Bed_Occupancy_Rate_Calc"].mean())
```

```python
data["ICU_Occupancy_Rate_Calc"] = data["ICU_Occupancy_Rate_Calc"].fillna(data["ICU_Occupancy_Rate_Calc"].mean())
```

```python
data["Staff_Utilization_Calc"] = data["Staff_Utilization_Calc"].fillna(data["Staff_Utilization_Calc"].mean())
```

```python
data["Readmission_Flag"] = data["Readmission_Flag"].fillna(data["Readmission_Flag"].mean())
```

```python
data["Transferred_Flag"] = data["Transferred_Flag"].fillna(data["Transferred_Flag"].mean())
```

```python
data["Equipment_InUse_Flag"] = data["Equipment_InUse_Flag"].fillna(data["Equipment_InUse_Flag"].mean())
```

```python
data["Transfer_Date"] = pd.to_datetime(data["Transfer_Date"], errors='coerce')
```
