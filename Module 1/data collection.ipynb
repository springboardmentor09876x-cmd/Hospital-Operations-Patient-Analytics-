```python
import pandas as pd

# Load datasets
my_df = pd.read_excel(r"C:\Users\thota\Downloads\Hospital_Dataset.xlsx")
reference_df = pd.read_csv(r"C:\Users\thota\Downloads\hospital_final_dataset(row_level).csv")

# Remove leading/trailing spaces from column names
my_df.columns = my_df.columns.str.strip()
reference_df.columns = reference_df.columns.str.strip()

# Print column names
print(my_df.columns.tolist())
print(reference_df.columns.tolist())
```

```python
# Find missing columns
missing_columns = [col for col in reference_df.columns if col not in my_df.columns]

# Merge missing columns into your dataset
my_df = my_df.merge(
    reference_df[['Patient ID'] + missing_columns],
    on='Patient ID',
    how='left'
)

# Save the merged dataset
my_df.to_excel("Hospital_Dataset_Merged.xlsx", index=False)

print("All missing columns have been added successfully!")
```
