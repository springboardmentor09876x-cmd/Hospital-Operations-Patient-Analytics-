```python
import pandas as pd
```

```python
synthetic_path = r"C:\Users\skkarishma\OneDrive\Apps\Synthetic_Hospital_Dataset.xlsx"

patient_path = r"C:\Users\skkarishma\Downloads\Patient_Operational_Raw_Dataset.xlsx"
```

```python
synthetic_df = pd.read_excel(synthetic_path)

patient_df = pd.read_excel(patient_path)
```

```python
print("Synthetic Dataset Columns:")
print(synthetic_df.columns.tolist())

print("\nPatient Operational Dataset Columns:")
print(patient_df.columns.tolist())
```

```python
common_columns = list(set(synthetic_df.columns) & set(patient_df.columns))

print("Common Columns:")
print(common_columns)
```

```python
merged_df = pd.concat(
    [synthetic_df, patient_df],
    axis=1
)
```

```python
merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]
```

```python
print(merged_df.head())

print()

print(merged_df.shape)

print()

print(merged_df.columns)
```

```python
required_columns = [
    "Hospital ID",
    "Hospital Number",
    "Hospital Type",
    "State",
    "City",
    "Department",
    "Department ID",
    "Doctor",
    "Nurses",
    "Staff Count",
    "Patient ID",
    "Patient Name",
    "Admission Date",
    "Gender",
    "Age",
    "Patient Analysis",
    "Patient Movement",
    "Patient Transfer",
    "Doctor In Charge",
    "Current Department Duration (Hours)",
    "Diagnosis",
    "Treatment",
    "Test Result",
    "Blood Type",
    "Length of Stay",
    "Room No",
    "Bed Available",
    "ICU Beds",
    "Bed Number",
    "Bed Occupied",
    "Equipment",
    "Equipment Usage Duration",
    "Insurance Provider",
    "Billing Amount",
    "Discharge Date",
    "Readmission"
]
```

```python
missing_columns = set(required_columns) - set(merged_df.columns)

print("Missing Columns:")
print(missing_columns)
```

```python
# Add missing columns

merged_df["Hospital Number"] = (
    "HOSP-" + merged_df["Hospital ID"].astype(str)
)

merged_df["Bed Available"] = (
    merged_df["Bed Occupied"].apply(
        lambda x: "Yes" if x == 0 else "No"
    )
    if "Bed Occupied" in merged_df.columns
    else "Unknown"
)

merged_df["Patient Transfer"] = "No"

merged_df["Equipment Usage Duration"] = (
    merged_df["Equipment"].apply(
        lambda x: "2-8 Hours" if pd.notna(x) else "0 Hours"
    )
)
```

```python
missing_columns = set(required_columns) - set(merged_df.columns)

print("Missing Columns:")
print(missing_columns)
```

```python
merged_df = merged_df[required_columns]
```

```python
print(merged_df.head())

print("\nTotal Columns:")
print(len(merged_df.columns))
```

```python
final_path = r"C:\Users\skkarishma\OneDrive\Documents\Final_Hospital_Dataset.csv"

merged_df.to_csv(
    final_path,
    index=False
)

print("Saved Successfully")
print(final_path)
```
