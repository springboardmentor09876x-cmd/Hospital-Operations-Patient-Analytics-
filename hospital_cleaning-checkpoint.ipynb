```python
hospital_df = hospital_df.drop_duplicates()
```

```python
hospital_df = hospital_df.dropna(subset=["Patient Name"])       #Remove Rows with Too Many Missing Values
```

```python
hospital_df = hospital_df.dropna(subset=["Age"])    #removes rows where essential data is missing.
```

```python
hospital_df["Age"] = hospital_df["Age"].fillna(hospital_df["Age"].median())   #Fill Missing Values
```

```python
hospital_df["Gender"] = hospital_df["Gender"].fillna("Unknown")                #Fill Missing Values for Gender
```

```python
hospital_df["Department"] = hospital_df["Department"].fillna("General")          #Fill missing values for Department
```

```python
hospital_df["Department"] = hospital_df["Department"].str.strip()           #Remove Extra Spaces (Trim)
```

```python
hospital_df["Department"] = hospital_df["Department"].replace({
    "CARDIOLOGY":"Cardiology",
    "cardiology":"Cardiology",
    "Cardio":"Cardiology"
})
```

```python
hospital_df["Department"] = hospital_df["Department"].replace({
    "Ortho":"Orthopedics",
    "ORTHOPEDICS":"Orthopedics"
})
```

```python
hospital_df["Age Group"] = pd.cut(
    hospital_df["Age"],
    bins=[0,18,35,60,100],
    labels=["Child","Young Adult","Adult","Senior"]
)
```
