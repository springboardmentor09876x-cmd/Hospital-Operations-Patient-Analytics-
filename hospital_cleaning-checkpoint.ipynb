```python
text_columns = hospital_df.select_dtypes(include="object").columns

for col in text_columns:
    hospital_df[col] = hospital_df[col].fillna("Unknown")
```

```python
numeric_columns = hospital_df.select_dtypes(include=["int64","float64"]).columns

for col in numeric_columns:
    hospital_df[col] = hospital_df[col].fillna(hospital_df[col].median())
```

```python
for col in text_columns:
    hospital_df[col] = hospital_df[col].str.strip()
```

```python
for col in text_columns:
    hospital_df[col] = hospital_df[col].str.title()
```

```python
hospital_df["Department"] = hospital_df["Department"].replace({
    "Er":"Emergency",
    "Emergency Room":"Emergency",
    "Cardio":"Cardiology",
    "Ortho":"Orthopedics",
    "Neuro":"Neurology"
})
```

```python
hospital_df[numeric_columns] = hospital_df[numeric_columns].apply(pd.to_numeric, errors="coerce")
```

```python
hospital_df[numeric_columns] = hospital_df[numeric_columns].fillna(
    hospital_df[numeric_columns].median()
)
```

```python
hospital_df["Admission Date"] = pd.to_datetime(
    hospital_df["Admission Date"],
    errors="coerce"
)

hospital_df["Discharge Date"] = pd.to_datetime(
    hospital_df["Discharge Date"],
    errors="coerce"
)
```

```python
hospital_df["Admission Date"] = hospital_df["Admission Date"].dt.strftime("%d-%m-%Y")

hospital_df["Discharge Date"] = hospital_df["Discharge Date"].dt.strftime("%d-%m-%Y")
```

```python
admission = pd.to_datetime(hospital_df["Admission Date"], dayfirst=True)

discharge = pd.to_datetime(hospital_df["Discharge Date"], dayfirst=True)
```

```python
hospital_df["Length of Stay"] = (
    discharge - admission
).dt.days
```

```python
hospital_df["Admission Date"] = pd.to_datetime(
    hospital_df["Admission Date"],
    dayfirst=True
)

hospital_df["Discharge Date"] = pd.to_datetime(
    hospital_df["Discharge Date"],
    dayfirst=True
)

hospital_df["Length of Stay"] = (
    hospital_df["Discharge Date"] -
    hospital_df["Admission Date"]
).dt.days
```
