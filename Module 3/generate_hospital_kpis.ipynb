```python
# importing libraries 
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\thota\Hospital-Operations-Patient-Analytics-\Module 2\hospital_cleaned.csv")   # or read_excel() if Excel
```

```python
# filling missing values in transfer date column 
df["Transfer_Date"] = df["Transfer_Date"].fillna("Not Transferred")
```

```python
# checking the column 
print(df["Transfer_Date"].dtype)

print(df["Transfer_Date"].head(10))

print(df["Transfer_Date"].isna().sum())
```

```python
# total admissions 
total_admissions = len(df)

df["Total_Admissions"] = total_admissions
```

```python
#Occupancy rate 
overall_occupancy = (
    df["Beds_Occupied_Count"].sum() /
    df["Dept_Bed_Capacity_Derived"].sum()
) * 100

df["Occupancy_Rate"] = round(overall_occupancy, 2)
```

```python
# Avreage length of stay 
average_los = df["Length of Stay"].mean()

df["Average_Length_of_Stay"] = round(average_los, 2)
```

```python
#Readmission Rate

readmission_rate = (
    df["Readmission_Flag"].sum() /
    len(df)
) * 100

df["Readmission_Rate"] = round(readmission_rate, 2)
```

```python
# Bed Utilization Rate
bed_utilization = (
    df["Beds_Occupied_Count"].sum() /
    df["Dept_Bed_Capacity_Derived"].sum()
) * 100

df["Bed_Utilization_Rate"] = round(bed_utilization, 2)
```

```python
# Department Efficiency Score
department_efficiency = (
    df["Staff_Utilization_Calc"].mean() +
    df["Bed_Occupancy_Rate_Calc"].mean()
) / 2

df["Department_Efficiency_Score"] = round(department_efficiency, 2)
```

```python
df.to_excel("hospital_final_dataset.xlsx", index=False)

print("KPI dataset saved successfully!")
```
