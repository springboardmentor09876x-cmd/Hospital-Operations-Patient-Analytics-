```python
import pandas as pd
from pathlib import Path

# -------------------------------
# File Paths
# -------------------------------
input_file = Path("data") / "hospital_raw_data.csv"
output_file = Path("data") / "hospital_cleaned.csv"

# -------------------------------
# Load Dataset
# -------------------------------
print("Loading Raw Dataset...")

df = pd.read_csv(input_file)

print("Dataset Loaded Successfully!")

# -------------------------------
# Remove Duplicate Rows
# -------------------------------
duplicates = df.duplicated().sum()
print(f"Duplicate Rows Found: {duplicates}")

df.drop_duplicates(inplace=True)

# -------------------------------
# Missing Values Before Cleaning
# -------------------------------
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

# -------------------------------
# Fill Numeric Columns
# -------------------------------
numeric_cols = df.select_dtypes(include=["number"]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# -------------------------------
# Fill Text Columns
# -------------------------------
text_cols = df.select_dtypes(include=["object"]).columns

for col in text_cols:
    if col not in [
        "Transfer_Date",
        "Transfer_From_Department",
        "Transfer_To_Department"
    ]:
        df[col] = df[col].fillna("Unknown")

# -------------------------------
# Handle Transfer Columns
# -------------------------------
df["Transfer_Date"] = df["Transfer_Date"].fillna("Not Transferred")
df["Transfer_From_Department"] = df["Transfer_From_Department"].fillna("Not Transferred")
df["Transfer_To_Department"] = df["Transfer_To_Department"].fillna("Not Transferred")

# -------------------------------
# Standardize Department Names
# -------------------------------
df["Department"] = (
    df["Department"]
      .astype(str)
      .str.strip()
      .str.title()
)

# -------------------------------
# Normalize Column Names
# -------------------------------
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

# -------------------------------
# Missing Values After Cleaning
# -------------------------------
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

missing_percent = (
    df.isnull().sum().sum()
    / (df.shape[0] * df.shape[1])
) * 100

print(f"\nOverall Missing Percentage: {missing_percent:.2f}%")

# -------------------------------
# Save Cleaned Dataset
# -------------------------------
df.to_csv(output_file, index=False)

print("\nCleaning Completed Successfully!")
print(f"Cleaned Dataset Saved To: {output_file}")
```
