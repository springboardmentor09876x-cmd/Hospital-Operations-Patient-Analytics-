```python
import pandas as pd
import numpy as np

## Load Raw Dataset
df = pd.read_csv(r"C:\Users\chere\Hospital_RawDataset_Final.csv")

df.head()

print(df.isnull().sum()[df.isnull().sum()>0])
```

```python
## Basic Dataset Information
print("Shape:", df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Information:")
df.info()

print("\nSummary Statistics:")
display(df.describe(include="all"))
```

```python
## Missing Value Analysis
missing = df.isnull().sum()

missing = missing[missing > 0]

print(missing)

missing_percentage = (
    df.isnull().sum() / len(df) * 100
).round(2)

print("\nMissing Percentage")

print(
    missing_percentage[
        missing_percentage > 0
    ]
)
```

```python
## Remove Duplicate Records
duplicates = df.duplicated().sum()

print("Duplicate Records:", duplicates)

df.drop_duplicates(inplace=True)

print("Dataset Shape:", df.shape)
```

```python
## Handle Missing Values
# Fill Insurance Provider

if "Insurance Provider" in df.columns:
    df["Insurance Provider"] = (
        df["Insurance Provider"]
        .fillna("Unknown")
    )

# Fill Transfer Destination

if "Transfer_To_Department" in df.columns:
    df["Transfer_To_Department"] = (
        df["Transfer_To_Department"]
        .fillna("Not Transferred")
    )

print(df.isnull().sum()[df.isnull().sum()>0])
```

```python
## Standardize Data
date_columns = [
    "Admission Date",
    "Discharge Date",
    "Transfer_Date"
]

for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(
            df[col],
            errors="coerce",
            dayfirst=True
        )

text_columns = [
    "Department",
    "Hospital Type",
    "Equipment Status",
    "Transferred",
    "Gender"
]

for col in text_columns:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.title()
        )
```

```python
## Data Validation
print("Negative Age:",
      (df["Age"] < 0).sum())

print("Negative Length of Stay:",
      (df["Length of Stay"] < 0).sum())

print("Discharge before Admission:",
      (df["Discharge Date"] < df["Admission Date"]).sum())

if {"Beds Occupied", "Beds Available"}.issubset(df.columns):
    print("Occupied > Available:",
          (df["Beds Occupied"] > df["Beds Available"]).sum())

print("Duplicate Rows:",
      df.duplicated().sum())
```

```python
## Export Cleaned Dataset
df.to_csv(
    "hospital_cleaned.csv",index=False
)
print("Cleaned dataset exported successfully.")
```
