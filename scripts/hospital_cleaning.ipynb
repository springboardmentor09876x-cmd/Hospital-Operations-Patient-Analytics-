```python
# Import pandas for data handling
import pandas as pd

# Define the data folder path 
path = r'C:\Users\pranv\OneDrive\Desktop\Hospital Operations & Patient Analytics\Hospital-Operations-Patient-Analytics-\data\\'

# Load Module 1's output dataset 
df_raw = pd.read_csv(path + 'hospital_raw_data.csv')

# Work on a copy of the data
df = df_raw.copy()

# Display  dataset
print("First 5 rows of the dataset:")
display(df.head())

print("\nShape of the dataset (rows, columns):")
print(df.shape)

print("\nColumn names:")
print(df.columns.tolist())

print("\nDataset info (data types, non-null counts):")
df.info()

print("\nData types of each column:")
print(df.dtypes)
```

```python
#  BASIC DATA VALIDATION (Before Cleaning)

# Check missing values in every column
print("Missing values per column:")
print(df.isnull().sum())

# Check how many duplicate rows exist
duplicate_count = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_count)

print("\nNumerical summary:")
display(df.describe())

# Categorical summary text columns
print("\nCategorical summary - Department:")
print(df['Department'].value_counts(dropna=False))

print("\nCategorical summary - Gender:")
print(df['Gender'].value_counts(dropna=False))

print("\nCategorical summary - Hospital_Type:")
print(df['Hospital_Type'].value_counts(dropna=False))

# Store key stats for later comparison
original_shape = df.shape
total_missing_before = df.isnull().sum().sum()
total_duplicates_before = df.duplicated().sum()

print("\n--- SUMMARY ---")
print("Original dataset shape:", original_shape)
print("Total missing values:", total_missing_before)
print("Total duplicate rows:", total_duplicates_before)
```

```python
# VERIFY AND CORRECT DATA TYPES


# Print data types before any conversion
print("Data types BEFORE conversion:")
print(df.dtypes)

# Convert date columns from text to actual datetime format
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'], errors='coerce')
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'], errors='coerce')

# Ensure Patient_ID and Hospital_ID stay as text 
df['Patient_ID'] = df['Patient_ID'].astype(str)
df['Hospital_ID'] = df['Hospital_ID'].astype(str)
df['Department_ID'] = df['Department_ID'].astype(str)
df['Equipment_ID'] = df['Equipment_ID'].astype(str)

# Ensure numeric columns are proper integers or floats
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Bed_Number'] = pd.to_numeric(df['Bed_Number'], errors='coerce')
df['Length_of_Stay'] = pd.to_numeric(df['Length_of_Stay'], errors='coerce')
df['Staff_Utilization'] = pd.to_numeric(df['Staff_Utilization'], errors='coerce')
df['Bed_Occupancy_Rate'] = pd.to_numeric(df['Bed_Occupancy_Rate'], errors='coerce')
df['ICU_Occupancy_Rate'] = pd.to_numeric(df['ICU_Occupancy_Rate'], errors='coerce')

# Print data types after conversion
print("\nData types AFTER conversion:")
print(df.dtypes)
```

```python
#  VALIDATE DATE COLUMNS

# Check for any invalid  dates after conversion
invalid_admission_dates = df['Admission_Date'].isnull().sum()
invalid_discharge_dates = df['Discharge_Date'].isnull().sum()

print("Invalid Admission_Date values:", invalid_admission_dates)
print("Invalid Discharge_Date values:", invalid_discharge_dates)

# Check if Admission_Date is ever AFTER Discharge_Date 
invalid_date_order = df[df['Admission_Date'] > df['Discharge_Date']]
print("\nNumber of records where Admission_Date is after Discharge_Date:", len(invalid_date_order))

# Recalculate Length_of_Stay directly from the validated dates
df['Length_of_Stay'] = (df['Discharge_Date'] - df['Admission_Date']).dt.days

# Handle invalid records (if Length_of_Stay is negative)
invalid_stay_count = (df['Length_of_Stay'] < 0).sum()
print("Number of records with negative Length_of_Stay:", invalid_stay_count)

# Fix: for any negative Length_of_Stay, set it to missing 
df.loc[df['Length_of_Stay'] < 0, 'Length_of_Stay'] = pd.NA

print("\nLength_of_Stay summary after validation:")
print(df['Length_of_Stay'].describe())
```

```python

# REMOVE DUPLICATE RECORDS

# Count duplicate rows before removal
duplicates_before = df.duplicated().sum()
print("Duplicate rows found:", duplicates_before)

# Store row count before removal
rows_before = df.shape[0]

# Remove duplicate rows 
df = df.drop_duplicates()

# Store row count after removal
rows_after = df.shape[0]

# Calculate how many rows were actually removed
rows_removed = rows_before - rows_after

print("Rows before removing duplicates:", rows_before)
print("Rows after removing duplicates:", rows_after)
print("Total duplicate rows removed:", rows_removed)
```

```python
#  HANDLE MISSING VALUES

# Print missing values before cleaning
missing_before = df.isnull().sum()
total_missing_before_cleaning = missing_before.sum()
print("Missing values per column BEFORE cleaning:")
print(missing_before)
print("\nTotal missing values BEFORE cleaning:", total_missing_before_cleaning)

# Separate columns into numeric and categorical  groups
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Fill missing numeric values using the median 
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value)
        print(f"Filled missing values in '{col}' with median: {median_value}")

# Fill missing categorical values using the mode 
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0]
        df[col] = df[col].fillna(mode_value)
        print(f"Filled missing values in '{col}' with mode: {mode_value}")

# Print missing values after cleaning
missing_after = df.isnull().sum()
total_missing_after_cleaning = missing_after.sum()
print("\nMissing values per column AFTER cleaning:")
print(missing_after)
print("\nTotal missing values AFTER cleaning:", total_missing_after_cleaning)

# Confirm we meet the <2% missing values requirement
missing_pct_after = (total_missing_after_cleaning / df.size) * 100
print(f"\nMissing value percentage after cleaning: {missing_pct_after:.2f}%")
```

```python
#  STANDARDIZE TEXT COLUMNS


# List of key text columns we want to standardize
text_columns_to_clean = [
    'Department', 'Hospital_Name', 'Doctor', 'Gender',
    'Blood Type', 'City', 'State', 'Hospital_Type',
    'Admission_Type', 'Test_Result'
]

# Only process columns that actually exist in our dataset
text_columns_to_clean = [col for col in text_columns_to_clean if col in df.columns]

print("Text columns being standardized:", text_columns_to_clean)

for col in text_columns_to_clean:
    #  Remove leading/trailing spaces
    df[col] = df[col].astype(str).str.strip()

    #  Replace multiple internal spaces with a single space
    df[col] = df[col].str.replace(r'\s+', ' ', regex=True)

    #  Convert to Title Case 
    df[col] = df[col].str.title()

    print(f"\nUnique values in '{col}' after standardization:")
    print(df[col].unique())
```

```python
#  NORMALIZE HEALTHCARE INDICATORS 

# Define a simple Min-Max normalization function
# Formula: (value - min) / (max - min) 
def min_max_normalize(column):
    return (column - column.min()) / (column.max() - column.min())

# List of KPI-related numeric columns to normalise
kpi_columns_to_normalize = [
    'Bed_Occupancy_Rate',
    'ICU_Occupancy_Rate',
    'Staff_Utilization',
    'Length_of_Stay'
]

# Only normalise columns that actually exist in our dataset
kpi_columns_to_normalize = [col for col in kpi_columns_to_normalize if col in df.columns]

print("KPI columns being normalized:", kpi_columns_to_normalize)

# Create new normalised columns 
for col in kpi_columns_to_normalize:
    new_col_name = col + '_Normalized'
    df[new_col_name] = min_max_normalize(df[col])
    print(f"\nCreated '{new_col_name}' | Min: {df[new_col_name].min()}, Max: {df[new_col_name].max()}")

# Preview the original vs normalized columns side by side
preview_cols = []
for col in kpi_columns_to_normalize:
    preview_cols.append(col)
    preview_cols.append(col + '_Normalized')

print("\nPreview of original vs normalized KPI columns:")
display(df[preview_cols].head())
```

```python
#  KPI VALIDATION FOR DASHBOARD READINESS

# List of columns that are critical for the Hospital Operations Dashboard
dashboard_critical_columns = [
    'Admission_Date', 'Discharge_Date', 'Department', 'Hospital_Name',
    'City', 'State', 'Bed_Occupancy_Rate', 'ICU_Occupancy_Rate',
    'Length_of_Stay', 'Staff_Utilization', 'Admission_Type'
]

# Only check columns that actually exist in our dataset
dashboard_critical_columns = [col for col in dashboard_critical_columns if col in df.columns]

print("Validating dashboard-critical columns:", dashboard_critical_columns)

# Check for any remaining missing values in these specific columns
print("\nMissing values in dashboard-critical columns:")
print(df[dashboard_critical_columns].isnull().sum())

# Check data types of these columns to confirm they're dashboard-ready
print("\nData types of dashboard-critical columns:")
print(df[dashboard_critical_columns].dtypes)

# Print summary statistics for the key numeric KPI columns
numeric_kpi_cols = ['Bed_Occupancy_Rate', 'ICU_Occupancy_Rate', 'Length_of_Stay', 'Staff_Utilization']
numeric_kpi_cols = [col for col in numeric_kpi_cols if col in df.columns]

print("\nSummary statistics for key KPI columns:")
display(df[numeric_kpi_cols].describe())

# Confirm no negative values exist in KPI columns 
for col in numeric_kpi_cols:
    negative_count = (df[col] < 0).sum()
    print(f"Negative values in '{col}': {negative_count}")
```

```python
#  FINAL DATASET VALIDATION

# Final dataset shape
final_shape = df.shape

# Final column names
final_columns = df.columns.tolist()

# Final data types
final_dtypes = df.dtypes

# Final missing values check
final_missing = df.isnull().sum().sum()

# Final duplicate check
final_duplicates = df.duplicated().sum()

print("===== FINAL DATASET VALIDATION =====")
print("\nFinal dataset shape:", final_shape)
print("\nFinal column names:")
print(final_columns)
print("\nFinal data types:")
print(final_dtypes)
print("\nTotal missing values (final):", final_missing)
print("Total duplicate rows (final):", final_duplicates)

print("\n===== BEFORE vs AFTER CLEANING SUMMARY =====")
print("Original dataset shape:", original_shape)
print("Number of duplicate rows removed:", rows_removed)
print("Missing values before cleaning:", total_missing_before_cleaning)
print("Missing values after cleaning:", final_missing)
print("Final dataset shape:", final_shape)

print("\nDataset is ready for Tableau Dashboard.")
```

```python
#  SAVE THE CLEANED DATASET

# Define the save path 
save_path = path + 'hospital_cleaned.csv'

# Save the cleaned dataset 
df.to_csv(save_path, index=False)

print("hospital_cleaned.csv saved successfully.")
print("Saved to:", save_path)

# Final confirmation checks
print("\n===== FINAL CONFIRMATION =====")
print("Rows in cleaned dataset:", df.shape[0])
print("Columns in cleaned dataset:", df.shape[1])
print("Missing values in final saved file:", df.isnull().sum().sum())
print("Duplicate rows in final saved file:", df.duplicated().sum())
print("\nModule 2 complete. hospital_cleaned.csv is ready for Tableau.")
```
