```python
import pandas as pd

hospital_df = pd.read_csv(r"C:\Users\skkarishma\.ipynb_checkpoints\hospital_cleaned-checkpoint.csv")
equipment_df = pd.read_csv(r"C:\Users\skkarishma\Downloads\equipment_flag(1) (1).csv")

hospital_df["Equipment Flag"] = equipment_df["Equipment Flag"]
```

```python
hospital_df.head()
```

```python
hospital_df["Readmission"] = hospital_df["Readmission"].map({
    "Yes": 1,
    "No": 0
})
```

```python
hospital_df["Readmission"].value_counts()
```

```python
total_admissions = hospital_df.shape[0]                     #Total Admissions
```

```python
hospital_df["Bed Occupied"] = hospital_df["Bed Occupied"].map({
    "Yes": 1,
    "No": 0
})                                                                               #Occupancy Rate
```

```python
occupied_beds = hospital_df["Bed Occupied"].sum()

total_beds = len(hospital_df)

occupancy_rate = (occupied_beds / total_beds) * 100
```

```python
average_los = hospital_df["Length of Stay"].mean()                         #Average Length of Stay
```

```python
hospital_df["Admission Date"] = pd.to_datetime(hospital_df["Admission Date"])
hospital_df["Discharge Date"] = pd.to_datetime(hospital_df["Discharge Date"])

hospital_df["Length of Stay"] = (
    hospital_df["Discharge Date"] -
    hospital_df["Admission Date"]
).dt.days

average_los = hospital_df["Length of Stay"].mean()
```

```python
hospital_df["Admission Date"] = pd.to_datetime(hospital_df["Admission Date"], format="%d-%m-%Y")
hospital_df["Discharge Date"] = pd.to_datetime(hospital_df["Discharge Date"], format="%d-%m-%Y")
```

```python
hospital_df["Admission Date"] = pd.to_datetime(
    hospital_df["Admission Date"],
    format="mixed",
    dayfirst=True
)

hospital_df["Discharge Date"] = pd.to_datetime(
    hospital_df["Discharge Date"],
    format="mixed",
    dayfirst=True
)
```

```python
hospital_df[["Admission Date", "Discharge Date"]].head()
```

```python
hospital_df["Admission Date"] = pd.to_datetime(
    hospital_df["Admission Date"],
    format="mixed",
    dayfirst=True
)

hospital_df["Discharge Date"] = pd.to_datetime(
    hospital_df["Discharge Date"],
    format="mixed",
    dayfirst=True
)

hospital_df["Admission Date"] = hospital_df["Admission Date"].dt.strftime("%d-%m-%Y")
hospital_df["Discharge Date"] = hospital_df["Discharge Date"].dt.strftime("%d-%m-%Y")
```

```python
hospital_df[["Admission Date", "Discharge Date"]].head(10)
```

```python
hospital_df["Admission Date"].isna().sum()
hospital_df["Discharge Date"].isna().sum()
```

```python
readmission_rate = (
    hospital_df["Readmission"].sum()
    / len(hospital_df)
) * 100                                                              #Readmission Rate
```

```python
bed_utilization_rate = (
    hospital_df["Bed Occupied"].sum()
    / len(hospital_df)
) * 100                                                           #Bed Utilization Rate
```

```python
department_efficiency = (
    hospital_df.groupby("Department")["Length of Stay"]
    .mean()
)

department_efficiency = (
    hospital_df.groupby("Department").size()
    / department_efficiency
)                                                                                 #Department Efficiency Score
```

```python
print("------ Hospital KPI Summary ------")

print("Total Admissions :", total_admissions)

print("Occupancy Rate :", round(occupancy_rate,2), "%")

print("Average Length of Stay :", round(average_los,2), "Days")

print("Readmission Rate :", round(readmission_rate,2), "%")

print("Bed Utilization Rate :", round(bed_utilization_rate,2), "%")

print("\nDepartment Efficiency Score")

print(department_efficiency)                                                         #KPI Calculation
```
